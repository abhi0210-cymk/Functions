```python
def sum_even_numbers(numbers):
    total = 0
    for num in numbers:
        if num % 2 == 0:
            total += num
    return total

# Example
print(sum_even_numbers([1, 2, 3, 4, 5, 6]))
```

```python
def reverse_string(s):
    return s[::-1]

# Example
print(reverse_string("Python"))
```

```python
def square_list(numbers):
    result = []
    for num in numbers:
        result.append(num ** 2)
    return result

# Example
print(square_list([1, 2, 3, 4, 5]))
```

```python
def is_prime(n):
    if n <= 1:   # 0 and 1 are not prime
        return False
    for i in range(2, int(n**0.5) + 1):  # check divisibility
        if n % i == 0:
            return False
    return True

# Test numbers from 1 to 200
for num in range(1, 201):
    if is_prime(num):
        print(num, "is Prime")
    else:
        print(num, "is Not Prime")
```

```python
class Fibonacci:
    def __init__(self, n_terms):
        self.n_terms = n_terms   # total terms to generate
        self.count = 0
        self.a, self.b = 0, 1    # first two Fibonacci numbers

    def __iter__(self):
        return self

    def __next__(self):
        if self.count < self.n_terms:
            if self.count == 0:
                self.count += 1
                return self.a
            elif self.count == 1:
                self.count += 1
                return self.b
            else:
                self.a, self.b = self.b, self.a + self.b
                self.count += 1
                return self.b
        else:
          raise StopIteration
          # Example usage
fib = Fibonacci(10)   # first 10 terms
for num in fib:
    print(num)
```

```python
def powers_of_two(n):
    for i in range(n + 1):
        yield 2 ** i

# Example usage
for val in powers_of_two(5):
    print(val)
```

```python
# Generator function that reads a file line by line
def read_file_line_by_line(filename):
    try:
        with open(filename, "r") as file:
            for line in file:
                yield line.strip()   # remove newline character
    except FileNotFoundError:
        # If file doesn't exist, create it with some sample content
        with open(filename, "w") as file:
            file.write("Hello World\n")
            file.write("This is a test file\n")
            file.write("Python Generators are cool!\n")
        # After creating, read again
        with open(filename, "r") as file:
            for line in file:
                yield line.strip()


# Example usage
if __name__ == "__main__":
    filename = "example.txt"
    for line in read_file_line_by_line(filename):
        print(line)
```

```python
# List of tuples
data = [(1, 5), (3, 1), (4, 7), (2, 3)]

# Sort based on second element of each tuple
sorted_data = sorted(data, key=lambda x: x[1])

print(sorted_data)
```

```python
# Celsius to Fahrenheit conversion function
c_to_f = lambda c: (c * 9/5) + 32

# List of temperatures in Celsius
celsius = [0, 20, 37, 100]

# Use map to convert each value
fahrenheit = list(map(c_to_f, celsius))

print("Celsius:   ", celsius)
print("Fahrenheit:", fahrenheit)
```

```python
# Function to check if a character is NOT a vowel
is_not_vowel = lambda ch: ch.lower() not in 'aeiou'

# Input string
text = "Hello World, Python is amazing!"

# Use filter() to remove vowels
result = "".join(filter(is_not_vowel, text))

print("Original:", text)
print("Without vowels:", result)
```
